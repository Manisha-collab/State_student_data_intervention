Generate a variable for the total_score (sum of reading, spelling, sentence
and arithmetic scores).


```python
#Here to get the total score we used the fuction sum 
```

```python
data1['Total_Score'] = data1[['spelling', 'sentence', 'arithmetic']].sum(axis=1)
```

```python
data1
```

 Create another variable for letter_grades based on the total_score. Pupils
should be assigned letter grades as follows: A: if total_score is between
80–98; B: if 60–79; C: if 40–59; D: if 20–39; F: if 0–19.

```python
import numpy as np
conditions = [
              (data1['Total_Score'] >=80)&(data1['Total_Score'] <=98),
              (data1['Total_Score'] >=60)&(data1['Total_Score'] <=79),
              (data1['Total_Score'] >=40)&(data1['Total_Score'] <=59),
              (data1['Total_Score'] >=20)&(data1['Total_Score'] <=39), 
              (data1['Total_Score'] >=0)&(data1['Total_Score'] <=19)
             ]
values = ['A','B','C','D','E']

data1['letter_grades']=np.select(conditions,values)

data1
```

Compute the average years of experience of teachers by school id. Name this
variable “avg_yrstaught”


```python
# to get the average years of experience by schoolid we we used Group by function and mean to get the average
```

```python
data2['avg_yrstaught'] = data2.groupby('schoolid')['yrstaught'].mean()
```
